```python
import cv2
import matplotlib.pyplot as plt
```

```python
imagePath = '../Dataset/Test Data/Img1090.jpg'
```

```python
image = cv2.imread(imagePath)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
RGB_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
RGB_img = cv2.cvtColor(gray, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
faces = faceCascade.detectMultiScale(
        gray,
        scaleFactor=1.5,
        minNeighbors=8,
        minSize=(5,5)
        )
```

```python
print("Found {0} faces".format(len(faces)))
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(image, (x,y), (x+w,y+h),(0,255,0), 2)
```

```python
status = cv2.imwrite('faces_detected_haar.jpg', image)
print("[INFO] Image faces_detected_haar.jpg written to filesystem: ", status)
```

```python
image = cv2.imread('faces_detected_haar.jpg')
RGB_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```
