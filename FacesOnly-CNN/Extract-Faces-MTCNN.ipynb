```python
import cv2
import matplotlib.pyplot as plt
from mtcnn import MTCNN
```

```python
imagePath = '../Dataset/Test Data/Img1090.jpg'
```

```python
image = cv2.imread(imagePath)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
RGB_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
RGB_img = cv2.cvtColor(gray, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
detector = MTCNN()
```

```python
faces = detector.detect_faces(RGB_img)
```

```python
for face in faces:
    print(face)
```

```python
print("Found {0} faces".format(len(faces)))
```

```python
for face in faces:
    (x,y,w,h) = face["box"]
    cv2.rectangle(image, (x,y), (x+w,y+h),(0,255,0), 2)
```

```python
status = cv2.imwrite('faces_detected_MTCNN.jpg', image)
print("[INFO] Image faces_detected_MTCNN.jpg written to filesystem: ", status)
```

```python
image = cv2.imread('faces_detected_MTCNN.jpg')
RGB_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```
